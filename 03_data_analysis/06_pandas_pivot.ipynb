## Сводная таблица

```python
# взаимосвязь между 
# постоянством клиентов (is_repeated_guest), 
# типом отеля (hotel) и отменой брони (is_canceled)
df.pivot_table(index=['is_repeated_guest'],
              columns=['hotel'],
              values='is_canceled')
```

```python
# взаимосвязь между 
# постоянством клиентов (is_repeated_guest),
# количеством детей (children)
# типом отеля (hotel) и отменой брони (is_canceled)
df.pivot_table(index=['is_repeated_guest'],
              columns=['hotel','children'],
              values='is_canceled')
```

```python
# взаимосвязь между 
# постоянством клиентов (is_repeated_guest),
# типом депозита (deposit_type)
# типом отеля (hotel) и отменой брони (is_canceled)
# + привести обобщенные значения по каждой группе  ("итоги")
df.pivot_table(index=['is_repeated_guest'],
              columns=['hotel','children'],
              values='is_canceled',
              aggfunc='mean')
```

```python
# используя функцию cut разбить значения 
# в столбце lead_time на 6 бинов (интервалов)
lead_time_intervals = pd.cut(df.lead_time, bins=6)
# вывести результат
lead_time_intervals
```

```python
# число вхождений в каждый бин (интервал)
pd.value_counts(lead_time_intervals)
```

```python
# взаимосвязь между 
# постоянством клиентов (is_repeated_guest),
# lead_time_intervals
# типом отеля (hotel) и отменой брони (is_canceled)
df.pivot_table(index=[lead_time_intervals],
              columns=['hotel','is_repeated_guest'],
              values='is_canceled',
              aggfunc='mean')
```
