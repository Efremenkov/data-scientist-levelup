## Подготовка данных

```python
# добавить новый столбец (=all_days), который является суммой stays_in_week_nights (=weekdays) и stays_in_weekend_nights (=weekends)
df['all_days'] = df.stays_in_week_nights + df.stays_in_weekend_nights
df
```

```python
# сгруппировать данные по lead_time (функция агрегации - среднее значение)
lead_time_groupby_result = df.groupby(['lead_time']).mean()

# сохранить в новую переменную сгруппированные значения для:
# all_days
all_days = lead_time_groupby_result.all_days
# weekdays
weekdays = lead_time_groupby_result.stays_in_week_nights
# weekends
weekends = lead_time_groupby_result.stays_in_weekend_nights
```

```python
# сгруппировать данные по arrival_date_month (функция агрегации - сумма)
arrival_date_month_groupby_result = df.groupby(['arrival_date_month']).sum()

# сохранить в новую переменную сгруппированные значения для is_canceled (=month_canceled)
month_canceled = arrival_date_month_groupby_result.is_canceled

# сохранить в новую переменную сгруппированные значения для is_canceled ТОЛЬКО для летных месяцев (=summer_canceled)
summer_canceled_df = pd.DataFrame(arrival_date_month_groupby_result.is_canceled[arrival_date_month_groupby_result.index.isin(['June', 'July', 'August'])])
month_ordered = pd.CategoricalDtype(['July', 'June', 'August'], ordered = True)
summer_canceled_df.index = summer_canceled_df.index.astype(month_ordered)
# summer_canceled.index
summer_canceled = summer_canceled_df.sort_index()
summer_canceled
```

## Графики (line plot)

### `Pyplot`-стиль

![](./images/09/line_plt.png)

```python
plt.plot(lead_time_groupby_result.index, all_days, label='all', color='green')
plt.plot(lead_time_groupby_result.index, weekdays, label='weekday', color='magenta')
plt.plot(lead_time_groupby_result.index, weekends, label='weekend', color='yellow')
plt.xlabel('lead time')
plt.ylabel('nights')
plt.title(label='Lead time and stay nights', loc='right')
plt.legend()
```

### ОО-стиль

**ВНИМАНИЕ**:
- Важно не "слепить" изображения (наименования для осей `y` должны быть хорошо видны).

![](./images/09/line_oo.png)

```python
fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(12, 3))
ax1.plot(lead_time_groupby_result.index, all_days, color='green')
ax2.plot(lead_time_groupby_result.index, weekdays, color='#bf00bf')
ax3.plot(lead_time_groupby_result.index, weekends, color='#bfbf00')
ax1.set_xlabel('lead time')
ax1.set_ylabel('nights (all)')
ax2.set_xlabel('lead time')
ax2.set_ylabel('nights (weekday)')
ax3.set_xlabel('lead time')
ax3.set_ylabel('nights (weekend)')
```

## Гистограммы (histogram)

### `Pyplot`-стиль

**ВНИМАНИЕ**:
- Различный уровень прозрачности (`alpha`).

![](./images/09/hist_plt.png)

```python
plt.hist(all_days, 
                           bins=10,  
                           orientation='vertical',
                           color='g', 
                           edgecolor='b',
                           linewidth=3,
                           alpha=0.5,
                           hatch="*",
                           linestyle="--", 
                           label='all')
plt.hist(weekdays, 
                           bins=10,  
                           orientation='vertical',
                           color='m', 
                           edgecolor='r',
                           linewidth=3,
                           alpha=0.5,
                           hatch=".",
                           linestyle="-.", 
                           label='weekday')
plt.hist(weekends, 
                           bins=10,  
                           orientation='vertical',
                           color='y', 
                           edgecolor='#4CE0B7',
                           linewidth=3,
                           alpha=0.5,
                           hatch="/",
                           linestyle=":", 
                           label='weekend')

plt.title(label='Stays nights', 
          loc='left')
plt.xticks(np.arange(0, 30, step=5))
plt.xlabel('lead time')
plt.ylabel('nights')

plt.legend(loc='center')
```

### ОО-стиль

**ВНИМАНИЕ**: 
- Различный уровень прозрачности (`alpha`).
- `ticks` должны быть одинаковыми для оси `x`.

![](./images/09/hist_oo.png)

```python
fig, ax = plt.subplots(2,1, figsize=(8, 10))

ax[0].hist(all_days, 
        bins=6, 
        orientation='vertical',
        color='g', 
        edgecolor='b',
        linewidth=1,
        alpha=0.5,
        hatch="*",
        linestyle="-", 
        label='nights (all)')

ax[1].hist(weekdays, 
        bins=10, 
        orientation='vertical',
        color='m', 
        edgecolor='r',
        linewidth=1,
        alpha=0.5,
        hatch=".",
        linestyle="-", 
        label='weekday')

ax[1].hist(weekends, 
        bins=10, 
        orientation='vertical',
        color='y', 
        edgecolor='#4CE0B7',
        linewidth=1,
        alpha=0.5,
        hatch="/",
        linestyle="-", 
        label='weekend')

ax[0].set_xlabel('lead time')
ax[0].set_ylabel('nights (all)')
ax[1].set_xlabel('lead time')
ax[1].set_ylabel('nights (weekend)')

ax[0].set_xticks(np.arange(0, 30, step=2.5))
ax[1].set_xticks(np.arange(0, 30, step=2.5))

ax[1].legend(loc='upper right')
```

## Круговые диаграммы (pie chart)

### `Pyplot`-стиль

**ВНИМАНИЕ**:
- Выделить сектор с **максимальным** процентом.

![](./images/09/pie_plt.png)

```python
month_canceled_df = pd.DataFrame(month_canceled)
M = max(month_canceled_df.is_canceled)

plt.pie(x=month_canceled_df.is_canceled, 
                                    explode=[0.125 if j == M else 0 for j in month_canceled_df.is_canceled], 
                                    labels=month_canceled.index, 
                                    autopct='%1.1f%%',
                                    shadow=True, 
                                    radius=1.4,
                                    textprops={'fontsize': 9})
plt.legend(title="Month",
           loc='best', 
           bbox_to_anchor=(1.9, 1),
           edgecolor='coral')
```

### ОО-стиль

**ВНИМАНИЕ**:
- Выделить сектор с **минимальным** процентом.
- Наименования секторов должны находится внутри.

![](./images/09/pie_oo.png)

```python
fig, ax = plt.subplots(figsize=(10, 10))

m = min(month_canceled_df.is_canceled)

ax.pie(x=month_canceled_df.is_canceled, 
       explode=[0.3 if j == m else 0 for j in month_canceled_df.is_canceled], 
       labels=month_canceled.index, 
       autopct='%1.1f%%',
       shadow=True, 
       startangle=60,
       labeldistance=0.6,
       pctdistance=0.5,
       textprops={'fontsize': 10})

ax.set_title("Canceled month")
```

## Ствол-лист диаграммы (stem plot)

### `Pyplot`-стиль

**ВНИМАНИЕ**:
- "Базовая линия" (`bottom`) - это медиана.

![](./images/09/stem_plt.png)

```python
plt.stem(month_canceled_df.index, month_canceled_df.is_canceled,
         label='median',
         linefmt="m-.",
         markerfmt="g*",
         basefmt='y--',
         bottom=month_canceled_df.is_canceled.median())

plt.xticks(rotation=45)

plt.legend(loc='best')
```

![](./images/09/stem_oo.png)

```python
fig, ax = plt.subplots(figsize=(8, 3))

ax.stem(month_canceled_df.index, month_canceled_df.is_canceled,
        label='median',
        linefmt="m-.",
        markerfmt="g*",
        basefmt='y--',
        bottom=month_canceled_df.is_canceled.median())


ax.xaxis.set_tick_params(rotation=75)
ax.yaxis.set_tick_params(rotation=-45)

ax.legend(loc='best')
```

## Ящик с усами (Box Plot)

### `Pyplot`-стиль

**ВНИМАНИЕ**:
- Особый стиль для отображения выбросов (outliers).

![](./images/09/box_plt.png)

```python
outliers = dict(marker='X',
                         markerfacecolor='y',
                         markeredgecolor='r',
                         markersize=9, 
                         alpha=0.6)

plt.boxplot(all_days, 
                notch=True, 
                vert=False,
                flierprops=outliers)

plt.title(label='Stay nights', 
          loc='center')
```

### ОО-стиль

**ВНИМАНИЕ**: 
- Особый стиль для отображения выбросов (outliers).

![](./images/09/box_oo.png)

```python
data = {'All': all_days, 'Weekday': weekdays, 'Weekend': weekends}

fig, ax = plt.subplots(figsize=(3, 7))

ax.boxplot(data.values(), 
           flierprops=outliers)

ax.set_xticklabels(data.keys())
```
